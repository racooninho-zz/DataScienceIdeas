```python
np.random.seed(101)

print(stocks.columns)
weights = np.array(np.random.random(len(stocks.columns)))

print('Random weights')
print(weights)

print('Rebalance after normalisation')

weights = weights/np.sum(weights) #normalisation
print(weights)

# Expected Return
print('Expected portfolio return')
expected_return = np.sum((log_returns.mean() * weights) * 252)
print(expected_return)
# log_returns.mean() is the daily return so we multiply by 252 
# business days to get annual

# expected volatility
print('Expected Volatility')
expected_volatility = np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*252,weights)))
#first we do this np.dot(log_returns.cov()*252,weights)
print(expected_volatility)

#Sharpe Ratio
print('Sharpe Ratio')
sharpe_ratio = expected_return / expected_volatility
print(sharpe_ratio)
```

```python
np.random.seed(101)

num_positions = 5001
all_weights = np.zeros((num_positions,len(stocks.columns)))
return_array = np.zeros(num_positions)
volatility_array = np.zeros(num_positions)
sharpe_ratio_array = np.zeros(num_positions)


for index in range(num_positions):

    weights = np.array(np.random.random(len(stocks.columns)))
    weights = weights/np.sum(weights) #normalisation
    
    #Save the weights
    
    all_weights[index,:] = weights
    
    # Expected Return
    return_array[index] = expected_return = np.sum(( log_returns.mean() * weights) * 252)

    # expected volatility
    volatility_array[index] = expected_volatility = np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*252,weights)))

    #Sharpe Ratio
    sharpe_ratio_array[index] = sharpe_ratio = expected_return / expected_volatility
```

<h1>Allocation with Optimization</h1>

```python
def get_return_volatility_sharperatio(weights):
    weights = np.array(weights)
    returns = np.sum(log_returns.mean() * weights) * 252
    volatility = np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*252,weights)))
    sharpe_ratio = returns/volatility
    return np.array([returns, volatility, sharpe_ratio])
```

```python
from scipy.optimize import minimize
```

```python
help(minimize)
```

```python
#for us to minimize we get the negative sharpe ratio
def negative_sharperatio(weights):
    return get_return_volatility_sharperatio(weights)[2] * -1
```

```python
def check_sum(weights):
    #return 0 if the of the weights is 1
    return np.sum(weights) - 1
```

<h2>Max volatility below</h2>

```python
volatility_value = 0.22

def max_volatility(weights):
    return get_return_volatility_sharperatio(weights)[1] - volatility_value

define_max_volatility = True
```

```python
if define_max_volatility == True:
    constraint = ({'type':'eq', 'fun':check_sum}
    #             , {'type':'eq', 'fun': lambda w: get_return_volatility_sharperatio(w)[1]-max_volatility}
                  ,{'type':'eq', 'fun':max_volatility}
                 )
else:
        constraint = ({'type':'eq', 'fun':check_sum}
    #             , {'type':'eq', 'fun': lambda w: get_return_volatility_sharperatio(w)[1]-max_volatility}
#                   ,{'type':'eq', 'fun':max_volatility}
                 )
```

```python
bounds = ((0,1),)

for i in range(0, len(stocks.columns)-1):
    bounds += (0,1),
#the weights for the allocation can only be between 0 and 1 (unless we short)
print(bounds)
print(len(stocks.columns))
```

```python
initial_allocation_guess = [0.25] * len(stocks.columns)
len(initial_allocation_guess)
```

```python
optiminal_results = minimize(negative_sharperatio, 
                             initial_allocation_guess, method='SLSQP',
                            bounds=bounds, constraints=constraint)
```

```python
#calcultate the efficient frontier
frontier_y = np.linspace(0,0.3,100)
```

```python
def minimize_volatility(weights):
    return get_return_volatility_sharperatio(weights)[1]
```

```python
frontier_volatility = []
#what is the best possible volatility for each return
for possible_return in frontier_y:
    constraint = ({'type':'eq', 'fun':check_sum},
                {'type':'eq', 'fun': lambda w: get_return_volatility_sharperatio(w)[0]-possible_return})

    
    result = minimize(minimize_volatility,initial_allocation_guess,method='SLSQP',
                            bounds=bounds, constraints=constraint)
    
    frontier_volatility.append(result['fun'])
```
